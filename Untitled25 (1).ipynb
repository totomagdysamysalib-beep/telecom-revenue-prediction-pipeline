```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
dataset['TotalCharges']=pd.to_numeric(dataset['TotalCharges'],errors='coerce')
dataset['TotalCharges']=dataset['TotalCharges'].fillna(0)

x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
columns_to_encode=[0,1,2,3,5,15,8,9, 10, 11, 12, 13]
for col in columns_to_encode:
  x_train[:,col]=le.fit_transform(x_train[:,col])
  x_test[:,col]=le.transform(x_test[:,col])
y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
```

```python
print(x_train[0])
```

```python
print(x_test)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[6, 7, 8,  14, 16])],remainder='passthrough')
x_train=ct.fit_transform(x_train)
x_test=ct.transform(x_test)
```

```python
print(x_train[0])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(C=0.1,solver='lbfgs')
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(y_pred,y_test))
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs']
}
grid_search=GridSearchCV(estimator=classifier,param_grid=parameters,cv=10,n_jobs=-1,scoring='accuracy')
grid_search.fit(x_train,y_train)
best_accuracy=grid_search.best_score_
best_parameters=grid_search.best_params_
print("The best accuracy {:.2f}%".format(best_accuracy*100))
print("The best parameters ",best_parameters)
```

```python
y_pred=grid_search.predict(x_test)
print(confusion_matrix(y_pred,y_test))
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(criterion= 'entropy', max_depth= 10, n_estimators= 100)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
parameters = {
'n_estimators': [50, 100],
    'max_depth': [None, 10],
    'criterion': ['gini', 'entropy']
}
grid_search=GridSearchCV(estimator=classifier,param_grid=parameters,cv=10,n_jobs=-1,scoring='accuracy')
grid_search.fit(x_train,y_train)
best_accuracy=grid_search.best_score_
best_parameters=grid_search.best_params_
print("The best accuracy {:.2f}%".format(best_accuracy*100))
print("The best parameters ",best_parameters)
```

```python
y_pred=grid_search.predict(x_test)
print(confusion_matrix(y_pred,y_test))
accuracy_score(y_pred,y_test)
```

```python
from xgboost import XGBClassifier
classifier=XGBClassifier(learning_rate= 0.1, max_depth= 3, n_estimators= 50)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(confusion_matrix(y_pred,y_test))
accuracy_score(y_pred,y_test)
```

```python
parameters = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 6]
}
grid_search=GridSearchCV(estimator=classifier,param_grid=parameters,cv=10,n_jobs=-1,scoring='accuracy')
grid_search.fit(x_train,y_train)
best_accuracy=grid_search.best_score_
best_parameters=grid_search.best_params_
print("The best accuracy {:.2f}%".format(best_accuracy*100))
print("The best parameters ",best_parameters)
```

```python
y_pred=grid_search.predict(x_test)
print(confusion_matrix(y_pred,y_test))
accuracy_score(y_pred,y_test)
```

```python
x=dataset.drop(columns=['customerID','TotalCharges']).values
y=dataset['TotalCharges'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
columns_to_encode=[0,1,2,3,5,15,8,9, 10, 11, 12, 13,18]
for col in columns_to_encode:
  x_train[:,col]=le.fit_transform(x_train[:,col])
  x_test[:,col]=le.transform(x_test[:,col])
```

```python
print(x_test[0])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[6, 7, 8,  14, 16])],remainder='passthrough')
x_train=ct.fit_transform(x_train)
x_test=ct.transform(x_test)
```

```python
print(x_test[0])
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from xgboost import XGBRegressor

regressor = XGBRegressor(n_estimators=200, learning_rate=0.1, random_state=0)

regressor.fit(x_train, y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
comparison_df=pd.DataFrame({"(Actual)":y_test,"(Predicted)":y_pred})
print(comparison_df.head(10))
```

```python
parameters = {
    'n_estimators': [50, 100,200],
    'learning_rate': [0.01, 0.1,0.2]

}
grid_search=GridSearchCV(estimator=regressor,param_grid=parameters,cv=10,n_jobs=-1,scoring='neg_mean_absolute_error')
grid_search.fit(x_train,y_train)
best_accuracy=grid_search.best_score_
best_parameters=grid_search.best_params_
print("The best accuracy {:.2f}".format(best_accuracy))
print("The best parameters ",best_parameters)
```

```python
y_pred=grid_search.predict(x_test)
comparison_df=pd.DataFrame({"(Actual)":y_test,"(Predicted)":y_pred})
print(comparison_df.head(10))
```
